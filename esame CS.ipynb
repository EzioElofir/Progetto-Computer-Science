#  12) Per ogni anno, calcolare l'importo totale dei prestiti. Ogni prestito che ha pianificato il tempo di scadenza e il tempo di erogazione in diversi anni deve avere il suo importo distribuito proporzionalmente al numero di giorni in ogni anno. Ad esempio, un prestito con tempo di erogazione del 1 ° dicembre 2016, scadenza prevista per il 30 gennaio 2018 e importo 5000 USD ha un importo di 5000 USD * 31 / (31 + 365 + 30) = 363,85 per il 2016, 5000USD * 365 / (31+ 365 + 30) = 4284.04 per il 2017 e 5000USD * 30 / (31 + 365 + 30) = 352.11 per il 2018.

```python
def day_calc(date1, date2):
    (begin,end) = (date1,date2) if date1<date2 else (date2,date1) #controllo  che le date nel dataframe siano in ordine
    date_list = list() #creo una lista con le date iniziali,finali e i valori tra le date
    date_list.append(begin)
    for y in range(begin.year+1, end.year+1):  
        date_list.append(pd.Timestamp(year=y, month=1, day=1, tz='utc')) 
    
    res = dict()
    for i in range(len(date_list)-1):  #iteriamo l'indice in base alla lunghezza della data 
        res[date_list[i].year] = (date_list[i+1] - date_list[i]).days # mettimi come chiavi tutti gli anni di date_list  come valore la differenza tra la data successiva a quella che considero e la data che considero
    return res #mi ritorna una tabella con una colonna (serie)

df_loans = df_loans.dropna() 
df_day_weighted = df_loans.apply(lambda x: day_calc(x['planned_expiration_time'],x['disburse_time']), axis = 1)
```

```python
df_day_weighted = pd.concat([df_day_weighted, df_loans['loan_id'], df_loans['loan_amount'], df_loans['duration'] ], axis = 1, sort = True)

df_day_weighted = df_day_weighted.rename({0: 'Years'}, axis = 1)
df_day_weighted
```

```python
##trasformo gli elementi all'interno di years in colonne
days = pd.concat([df_day_weighted.drop(['Years'], axis=1), df_day_weighted['Years'].apply(pd.Series)], axis=1) 
days.fillna(0, inplace = True) #riempio le posizione na con gli 0
days.head()
```

```python
amount = days[[2011,2012,2013,2014,2015,2016,2017,2018]].div(days.duration, axis=0).mul(days.loan_amount, axis = 0) #applichiamo la formula proposta nell'esercizio
amount
```

```python
amounts = pd.concat([days['loan_id'], amount],  axis = 1 )
#utilizziamo come indice loan id, con melt faccio una trasposizione
amounts = amounts.melt(id_vars=["loan_id"], var_name="Year", value_name="amount_per_year")  

amounts.groupby(['loan_id', 'Year']).sum()
```
